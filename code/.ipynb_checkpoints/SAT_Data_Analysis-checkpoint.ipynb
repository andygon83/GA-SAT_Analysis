##### 1. What does the data describe?

The data recorded the average scores and participation rates for students. The events are

##### 2. Does the data look complete? Are there any obvious issues with the observations?

The data is complete. Upon inspection I also realized that the last list in the file holds the

##### 3. Describe in words what each variable(column) is.

First, I will start with the "States" column. It acts as a index to the data since it is the 
only immutable series we will work with. This column is populated by state name abbreviations

The second column is the rate of participation by students in each state. Though not explicitly mentioned, you get this number by dividing the number of students taking the SAT

The final column operates the same as the aforementioned column, but using the math section of 
the SAT.

## Dive into the data.

```python
import csv
with open('../data/sat_scores.csv') as data:
    reader = csv.reader(data)
    satdata = list(reader)
```

```python
print satdata
```

##### The data appear complete in it's format and it includes a summary entry as the last list

```python
header = satdata[0]
sat_data = satdata[1:-1]
print header
del sat_data[27]  # I'm deleting Ohio after noticing a problem in the data in later analysis
sat_data[:5]
```

##### I now make a list with the states names for later use

```python
states = []
for i in sat_data:
    states.append(i[0])
```

##### and take a look at the types of data I'm using since imports don't always come as integers or floats.

```python
print type(sat_data[0][0]), type(sat_data[1][0]), type(sat_data[2][0]), type(sat_data[3][0])
```

##### I do a quick conversion to integers using list comprehension, and we're ready for analysis

```python
isat_data = [[int(j) for j in i[1:4]] for i in sat_data[:-1]]
```

## Analysis

##### I begin by converting the data into dictionaries that map the state name with each individual metric for that state.

```python
rate_col = dict(zip(states, [i[0] for i in isat_data]))

verbal_col = dict(zip(states, [i[1] for i in isat_data]))

math_col = dict(zip(states, [i[2] for i in isat_data]))


pretotal = [[i[0],i[1],i[2]] for i in isat_data]
total = dict(zip(states, pretotal))
```

##### And compiling it into a single dictionary based on variable

```python
maindict = {'math':[], 'verbal':[], 'rate':[]}
for i in isat_data:
    maindict['rate'].append(i[0])
    maindict['verbal'].append(i[1])
    maindict['math'].append(i[2])
```

##### We look at the limits of the data printing their respective mins and maxs

```python
print "Rate Column Min = " + str(min(maindict['rate'])) + ", Max = "+str(max(maindict['rate']))
print "Verbal Column Min = "+str(min(maindict['verbal']))+", Max = "+str(max(maindict['verbal']))
print "Math Column Min = "+str(min(maindict['math']))+", Max = "+str(max(maindict['math']))
```

##### and their standard deviations, computed using list comprehesion since the limits of this exercise prevent us from using loops or pandas library. I tried to be efficient by writing a function instead.

```python
def stdcol(column):
    return ((sum([y ** 2 for y in [x-(sum(maindict[column])/len(maindict[column])) for x in maindict[column]]])/len(maindict[column]))**0.5)


print "The standard deviation of the rate column is "+str(stdcol('rate'))
print "The standard deviation of the avg verbal score column is "+str(stdcol('verbal'))
print "The standard deviation of the avg math score column is "+str(stdcol('math'))
```

## Visualizing the data

```python
import matplotlib.pyplot as plt
%config InlineBackend.figure_format = 'retina'

plt.style.use('bmh')
plt.hist(maindict['rate'], ec='black', color= 'orangered', bins=17, alpha = 0.8)
plt.xlabel('% of high school students')
plt.ylabel('Number of states')
plt.title('Student Participation in SATs', fontsize = 'x-large')
plt.show()
```

```python
plt.hist(maindict['math'], ec='black', color= 'maroon', bins=20, alpha = 0.8)
plt.xlabel('Score ranges')
plt.ylabel('Frequency (in states)')
plt.title('SAT Math Avg Score', fontsize = 'x-large')
plt.show()
```

```python

plt.hist(maindict['verbal'], ec='black', color= 'olive', bins=22, alpha = 0.8)
plt.xlabel('Score ranges')
plt.ylabel('Frequency (in states)')
plt.title('SAT Verbal Avg Score', fontsize = 'x-large')
plt.show()
```

##### Scatter plot to observe relationship between variables

```python
x = [i[1] for i in total.values()]
y = [i[2] for i in total.values()]
labels = [i for i in total]
```

```python

plt.scatter(x , y, alpha=0.5)
plt.title('Mapping performance')
plt.xlabel('Verbal Scores')
plt.ylabel('Math Scores')
for labels, x, y in zip(labels, x, y):
    plt.annotate(
        labels,
        xy=(x, y), xytext=(-20, 20),
        textcoords='offset points', ha='right', va='bottom',
        bbox=dict(boxstyle='round,pad=0.5', fc='white', alpha=0.5),
        arrowprops=dict(arrowstyle = '->', connectionstyle='arc3,rad=0', color='black'))

plt.show()
```

Overall, what we see is that states with low participation are scoring very high. Upon research, I found that states that seem to relate high scores to low participation are due to the following factors:

1) Those states have the highest presence of their main competitor: ACT. This means that students are not as compelled to take the SAT to secure a university spot. 

2) Those states are predominantly agricultural and with low density population centers. Meaning that the local job market does not emphasize a college education, thus less impetus to participate in the exam.

3) The 'coasts' have a higher representation of participation, along with a higher representation of immigrants and higher population density.
